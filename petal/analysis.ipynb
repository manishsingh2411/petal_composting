```python
import json

json_file_path = r"C:\Users\manish.singh\OneDrive - METRO\Documents\cursor\cursor_explore\petal\analysis_data\logs_result.json"

with open(json_file_path, "r", encoding="utf-8") as f:
    logs_data = json.load(f)

import pandas as pd

df = pd.DataFrame(logs_data)
# Display top-level columns
display(df.head())

# Parse embedded JSON in the 'message' column, if available
def parse_message(msg):
    try:
        return json.loads(msg)
    except Exception:
        return {}

message_df = df['message'].apply(parse_message).apply(pd.Series)

# Merge back to main DataFrame for easier analysis
merged_df = pd.concat([df, message_df], axis=1)

# Filter to only log entries where some contact info (phone/email) is present
leads_df = merged_df[(merged_df['phone'].astype(str).str.len() > 0) | (merged_df['email'].astype(str).str.len() > 0)]

# Quick summary statistics
print("Total log records:", len(merged_df))
print("Number of unique sessions:", merged_df['sessionId'].nunique())
print("Number of lead submissions with phone or email:", len(leads_df))
if len(leads_df):
    display(leads_df[['ts', 'eventType', 'section', 'phone', 'email']])

# Analyze which sections are most engaged before a lead is captured
if len(leads_df):
    lead_sections = leads_df['section'].value_counts()
    print("Lead section breakdown:")
    print(lead_sections)
else:
    print("No leads (phone/email) found in current logs.")

# Analyze historical lead generation vs. time (if applicable)
if 'ts' in leads_df:
    leads_df['datetime'] = pd.to_datetime(leads_df['ts'], unit='ms')
    leads_per_day = leads_df.groupby(leads_df['datetime'].dt.date).size()
    print("Leads per day:")
    leads_per_day
```
